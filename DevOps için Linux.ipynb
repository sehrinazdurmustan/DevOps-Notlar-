## Başlangıç: Terminal ve Shell



### � Terminal Nedir?
Terminal (konsol), Linux sistemle etkileşime girdiğimiz metin tabanlı arayüzdür.

### 🐚 Shell Nedir?
Shell, kullanıcı komutlarını işletim sistemine ileten komut yorumlayıcısıdır.

### ⌨️ Terminal Bilgileri
- Prompt görünümü: `[kullanici@hostname ~]$`
- `$`: Normal kullanıcı
- `#`: Root (yönetici) kullanıcı
- `~`: Kullanıcının ev dizini

### ❗ Önemli Noktalar
- Linux büyük/küçük harfe duyarlıdır
- Komutlar boşlukla ayrılır
- Tab tuşu otomatik tamamlama yapar
- Yukarı/aşağı ok tuşları komut geçmişini gösterir

# : Temel Komutlar 🌟

## 1.1 Dizin Gezinme Komutları

### pwd (Print Working Directory)
- Mevcut çalışma dizinini gösterir
- Kullanımı: `pwd`

### cd (Change Directory)
- Dizinler arası geçiş yapar
- `cd dizin`: Belirtilen dizine git
- `cd ..`: Üst dizine git
- `cd ~`: Ana dizine git
- `cd -`: Önceki dizine dön

### ls (List)
- Dizin içeriğini listeler
- `ls`: Basit liste
- `ls -l`: Detaylı liste
- `ls -a`: Gizli dosyaları göster
- `ls -h`: İnsan okunabilir boyutlar

## 1.2 Temel Dosya Görüntüleme

### cat (Concatenate)
- Dosya içeriğini gösterir
- `cat dosya`: İçeriği göster
- `cat > dosya`: Yeni içerik gir
- `cat >> dosya`: İçeriğe ekle

### less
- Uzun dosyaları sayfa sayfa gösterir
- Space: Sonraki sayfa
- b: Önceki sayfa
- q: Çıkış

## 1.3 Yardım Alma

### man (Manual)
- Komutların kılavuzunu gösterir
- `man komut`: Komut kılavuzu
- `man -k kelime`: Anahtar kelime ara

### help
- Komut yardımı gösterir
- `komut --help`: Yardım göster

````python
# Temel komut örnekleri
[user@hostname ~]$ pwd
/home/user

# Dizin gezinme
[user@hostname ~]$ cd Documents
[user@hostname Documents]$ pwd
/home/user/Documents
[user@hostname Documents]$ cd ..
[user@hostname ~]$ pwd
/home/user

# Dizin listeleme
[user@hostname ~]$ ls
Documents  Downloads  Pictures  Videos
[user@hostname ~]$ ls -l
total 16
drwxr-xr-x 2 user user 4096 Sep 22 10:00 Documents
drwxr-xr-x 2 user user 4096 Sep 22 10:00 Downloads
drwxr-xr-x 2 user user 4096 Sep 22 10:00 Pictures
drwxr-xr-x 2 user user 4096 Sep 22 10:00 Videos

# Dosya görüntüleme
[user@hostname ~]$ echo "Merhaba Linux!" > test.txt
[user@hostname ~]$ cat test.txt
Merhaba Linux!

# Yardım görüntüleme
[user@hostname ~]$ man ls
# (Man sayfası açılır)
[user@hostname ~]$ ls --help
Usage: ls [OPTION]... [FILE]...
List information about the FILEs...
...
````

# 2: Dosya ve Dizin İşlemleri 📂

## 2.1 Dosya Oluşturma ve Düzenleme

### touch
- Boş dosya oluşturur
- Dosya zaman damgasını günceller
- `touch dosya`: Yeni dosya oluştur

### nano/vim
- Metin editörleri
- `nano dosya`: Dosya düzenle
- Ctrl+O: Kaydet
- Ctrl+X: Çık

### echo
- Metin yazdırma ve dosyaya yazma
- `echo "metin"`: Ekrana yazdır
- `echo "metin" > dosya`: Dosyaya yaz
- `echo "metin" >> dosya`: Dosyaya ekle

## 2.2 Dosya Kopyalama ve Taşıma

### cp (Copy)
- `cp kaynak hedef`: Dosya kopyala
- `cp -r kaynak hedef`: Dizin kopyala
- `cp -i`: Üzerine yazmadan önce sor

### mv (Move)
- Dosya/dizin taşır veya yeniden adlandırır
- `mv eski yeni`: Yeniden adlandır
- `mv dosya dizin/`: Taşı

## 2.3 Dosya Silme

### rm (Remove)
- `rm dosya`: Dosya sil
- `rm -r dizin`: Dizin sil
- `rm -i`: Silmeden önce sor
- `rm -f`: Zorla sil (dikkatli kullanın!)

## 2.4 Dizin İşlemleri

### mkdir (Make Directory)
- `mkdir dizin`: Dizin oluştur
- `mkdir -p a/b/c`: İç içe dizinler oluştur

### rmdir (Remove Directory)
- `rmdir dizin`: Boş dizin sil

````python
# Dosya oluşturma ve düzenleme
[user@hostname ~]$ touch yeni.txt
[user@hostname ~]$ echo "İlk satır" > yeni.txt
[user@hostname ~]$ echo "İkinci satır" >> yeni.txt
[user@hostname ~]$ cat yeni.txt
İlk satır
İkinci satır

# Dosya kopyalama ve taşıma
[user@hostname ~]$ cp yeni.txt yedek.txt
[user@hostname ~]$ mkdir yedekler
[user@hostname ~]$ mv yedek.txt yedekler/
[user@hostname ~]$ ls yedekler/
yedek.txt

# Dizin işlemleri
[user@hostname ~]$ mkdir -p proje/src/main
[user@hostname ~]$ tree proje/
proje/
└── src/
    └── main/

# Dosya silme
[user@hostname ~]$ rm yeni.txt
[user@hostname ~]$ rm -r yedekler/
[user@hostname ~]$ ls
proje/
````

#  3: Gelişmiş Metin İşleme 📝

## 3.1 Metin Arama ve Filtreleme

### grep (Global Regular Expression Print)
- Metin içinde arama yapar
- `grep "kelime" dosya`: Kelime ara
- `grep -i "KELIME" dosya`: Büyük/küçük harf duyarsız
- `grep -r "kelime" dizin`: Alt dizinlerde ara
- `grep -v "kelime" dosya`: Eşleşmeyenleri göster

### find ve locate
- `find dizin -name "dosya"`: Dosya ara
- `find . -type f`: Sadece dosyaları listele
- `locate dosya`: Sistem genelinde hızlı arama

## 3.2 Metin İşleme ve Sıralama

### sort
- Metin sıralama
- `sort dosya`: Alfabetik sırala
- `sort -r dosya`: Ters sırala
- `sort -n dosya`: Sayısal sırala

### uniq
- Tekrar eden satırları işler
- `uniq dosya`: Tekrarları kaldır
- `uniq -c`: Tekrar sayısını göster
- `uniq -d`: Sadece tekrar edenleri göster

## 3.3 Metin Analizi

### wc (Word Count)
- `wc dosya`: Satır, kelime ve karakter sayısı
- `wc -l`: Sadece satır sayısı
- `wc -w`: Sadece kelime sayısı

### head ve tail
- `head dosya`: İlk 10 satır
- `head -n 5`: İlk 5 satır
- `tail dosya`: Son 10 satır
- `tail -f`: Canlı takip

## 3.4 Pipe (|) Kullanımı
Komutları birleştirerek güçlü işlemler yapabilirsiniz:
- `ls -l | grep "pdf"`: PDF dosyalarını listele
- `sort dosya.txt | uniq -c`: Tekrar sayılarıyla listele
- `ps aux | grep "chrome"`: Chrome süreçlerini bul

````python
# Metin arama örnekleri
[user@hostname ~]$ echo -e "Linux\nWindows\nMacOS\nLinux" > isletim.txt

# Grep kullanımı
[user@hostname ~]$ grep "Linux" isletim.txt
Linux
Linux
[user@hostname ~]$ grep -c "Linux" isletim.txt
2

# Sort ve uniq kullanımı
[user@hostname ~]$ sort isletim.txt
Linux
Linux
MacOS
Windows
[user@hostname ~]$ sort isletim.txt | uniq -c
      2 Linux
      1 MacOS
      1 Windows

# Metin analizi
[user@hostname ~]$ wc isletim.txt
 4  4 24 isletim.txt  # satır kelime karakter

# Head ve tail
[user@hostname ~]$ head -n 2 isletim.txt
Linux
Windows
[user@hostname ~]$ tail -n 2 isletim.txt
MacOS
Linux

# Pipe kullanım örnekleri
[user@hostname ~]$ ls -l | grep "txt"
-rw-r--r-- 1 user user 24 Sep 22 10:00 isletim.txt
[user@hostname ~]$ cat isletim.txt | sort | uniq
Linux
MacOS
Windows
````

````python
# Dosya sistemi inceleme
[user@hostname ~]$ ls -l /
total 24
dr-xr-xr-x   2 root root 4096 Sep 22 10:00 bin
drwxr-xr-x   4 root root 4096 Sep 22 10:00 etc
drwxr-xr-x   4 root root 4096 Sep 22 10:00 home
drwxr-xr-x   2 root root 4096 Sep 22 10:00 var
...

# Ortam değişkenleri
[user@hostname ~]$ echo $USER
user
[user@hostname ~]$ echo $HOME
/home/user
[user@hostname ~]$ echo $PATH
/usr/local/bin:/usr/bin:/bin:/usr/sbin:/sbin

# Yeni ortam değişkeni tanımlama
[user@hostname ~]$ export TEST="Merhaba"
[user@hostname ~]$ echo $TEST
Merhaba

# Kullanıcı yönetimi (root olarak)
[root@hostname ~]# useradd yenikullanici
[root@hostname ~]# passwd yenikullanici
New password: 
Retype new password: 
passwd: password updated successfully

# Kullanıcı grupları
[root@hostname ~]# groups yenikullanici
yenikullanici : yenikullanici

# Dosya izinleri
[user@hostname ~]$ ls -l test.txt
-rw-r--r-- 1 user user 0 Sep 22 10:00 test.txt
[user@hostname ~]$ chmod 755 test.txt
[user@hostname ~]$ ls -l test.txt
-rwxr-xr-x 1 user user 0 Sep 22 10:00 test.txt

# Kullanıcı bilgileri
[user@hostname ~]$ grep user /etc/passwd
user:x:1000:1000:User:/home/user:/bin/bash
````

## Dosya İzinleri Detaylı Açıklama

### İzin Yapısı
```
-rwxr-xr-x
│└┬┘└┬┘└┬┘
│ │  │  └── Diğerleri için izinler (Others)
│ │  └───── Grup için izinler (Group)
│ └──────── Sahibi için izinler (User)
└────────── Dosya tipi (- dosya, d dizin, l link)
```

### İzin Tipleri
- `r` (4): Okuma izni
- `w` (2): Yazma izni
- `x` (1): Çalıştırma izni

### Örnekler
- `chmod 755`: rwxr-xr-x (Sahibi: tam yetki, Diğerleri: okuma ve çalıştırma)
- `chmod 644`: rw-r--r-- (Sahibi: okuma/yazma, Diğerleri: sadece okuma)
- `chmod 700`: rwx------ (Sadece sahibi tam yetki)

### Özel İzinler
- `SUID (4000)`: Dosya sahibinin yetkileriyle çalıştır
- `SGID (2000)`: Grup yetkileriyle çalıştır
- `Sticky Bit (1000)`: Sadece sahibi silebilir

### Sembolik Gösterim
- `u`: Kullanıcı (user)
- `g`: Grup (group)
- `o`: Diğerleri (others)
- `a`: Tümü (all)
- `+`: İzin ekle
- `-`: İzin kaldır
- `=`: İzin ayarla

Örnek: `chmod u+x dosya` (Sahibine çalıştırma izni ekle)

# Ek Linux Komutları ve İleri Düzey Kullanım

## 1. Alias ve Kısayollar

### Alias Kullanımı
- `alias`: Mevcut alias'ları listele
- `alias ls='ls --color=auto'`: Yeni alias tanımla
- `unalias ls`: Alias'ı kaldır
- `alias ll='ls -la'`: Uzun liste için kısayol

## 2. Gelişmiş Dosya İşlemleri

### tee Komutu
- Çıktıyı hem ekrana hem dosyaya yazar
- `ls | tee cikti.txt`: Liste hem ekranda görünür hem dosyaya yazılır
- `ls -l | tee -a cikti.txt`: Dosyaya ekleme modunda yazar

### cat'in Farklı Kullanımları
- `cat > dosya`: Yeni içerik oluştur (Ctrl+D ile çık)
- `cat >> dosya`: Var olan içeriğe ekle
- `cat dosya1 dosya2 > birlesik`: Dosyaları birleştir

## 3. Pipe ve Filtreleme İşlemleri

### Çoklu Pipe Örnekleri
```bash
ls -l /etc | nl                    # Satır numarası ekle
ls -l /etc | tail | cat -n         # Son satırları numaralandır
ls -l /etc | grep "conf" | sort    # conf içeren satırları sırala
```

### Metin İşleme Kombinasyonları
```bash
sort dosya.txt | uniq              # Tekrarları kaldır
sort dosya.txt | uniq -c           # Tekrar sayısını göster
sort -r /etc/passwd | head         # Ters sıralı ilk 10 satır
grep -i "ROOT" /etc/passwd         # Büyük/küçük harf duyarsız ara
```

## 4. Dosya Arama ve Konumlandırma

### locate ve whereis
- `locate dosya`: Sistem genelinde hızlı arama
- `whereis komut`: Komut konumunu ve man sayfasını bul
- `which komut`: Komutun tam yolunu göster

### find ile Detaylı Arama
```bash
find . -name "*.txt"               # .txt dosyalarını bul
find . -type f -size +1M           # 1MB'dan büyük dosyalar
find . -mtime -7                   # Son 7 günde değişenler
```

````python
# Alias örnekleri
[user@hostname ~]$ alias
alias ll='ls -la'
alias ls='ls --color=auto'
[user@hostname ~]$ alias dir='ls -l'
[user@hostname ~]$ dir
total 4
drwxr-xr-x 2 user user 4096 Sep 22 10:00 Documents

# tee kullanımı
[user@hostname ~]$ ls -l | tee liste.txt
total 4
drwxr-xr-x 2 user user 4096 Sep 22 10:00 Documents
[user@hostname ~]$ cat liste.txt
total 4
drwxr-xr-x 2 user user 4096 Sep 22 10:00 Documents

# Cat ile dosya işlemleri
[user@hostname ~]$ echo "İlk satır" > test.txt
[user@hostname ~]$ echo "İkinci satır" >> test.txt
[user@hostname ~]$ cat test.txt
İlk satır
İkinci satır

# Pipe örnekleri
[user@hostname ~]$ ls -l /etc | nl
     1	total 1048
     2	drwxr-xr-x 2 root root 4096 Sep 22 10:00 acpi
     3	drwxr-xr-x 2 root root 4096 Sep 22 10:00 alternatives

# Metin işleme
[user@hostname ~]$ echo -e "elma\nelma\narmut" > meyveler.txt
[user@hostname ~]$ sort meyveler.txt | uniq -c
      2 elma
      1 armut

# Dosya arama
[user@hostname ~]$ whereis ls
ls: /usr/bin/ls /usr/share/man/man1/ls.1.gz
[user@hostname ~]$ locate bash | head -3
/etc/bash.bashrc
/etc/bash.bash_logout
/etc/skel/.bash_logout

# find örnekleri
[user@hostname ~]$ find . -name "*.txt"
./test.txt
./liste.txt
./meyveler.txt
````
